#Transformada de Fourier Continua 

<center>
$Función:~x(t)=t,~0<t<1$

$x(f)=\dfrac{1}{τ}\int_τ{x(t)}(e^{-iπft})dt$

$\tau=1$

$x(f)=\int_0^1te^{-iπft}dt$



$=\dfrac{i}{2\pi f }\left|te^{-i2\pi ft}\right|_0^1+\dfrac{1}{i2\pi f }\int_0^1{e^{-i2\pi ft}}$

$=\dfrac{i}{2\pi f }e^{-i2\pi f}+
\dfrac{1}{(2\pi f )^2}\left|e^{-i2\pi ft}\right|_0^1$

$=\dfrac{i}{2\pi f }e^{-i2\pi f}+
\dfrac{1}{(2\pi f )^2}(e^{-2i \pi f}-1)$

$=\dfrac{i}{2\pi f }(cos(2 \pi f)-i~sin(2 \pi f))+
\dfrac{1}{(2\pi f )^2}(cos(2 \pi f)-i~sin(2 \pi f)-1)$

$\boxed{=\dfrac{1}{2\pi f }(\dfrac{1}{2 \pi f}(cos(2 \pi f)-1)+sin(2 \pi f))+\dfrac{i}{2\pi f }(cos(2 \pi f)-\dfrac{sin(2 \pi f)}{2 \pi f})}$


como $c_k=\dfrac{a_k}{2}-i\dfrac{b_k}{2}$


entonces:

$a_k=2Re(c_k)$

$b_k=-2Im(c_k)$


</center>



```python
# =============================================================================
#                      Librerías 
# =============================================================================
#Librerías con las que se trabajara:
import numpy as np
import matplotlib.pyplot as plt
import sys

# =============================================================================
#                      DEFINIR FUNCIÓN 
# =============================================================================
#
#     
#x(t)=t, para 0<t<1
#     
#
#Incremento de tiempo 
dt = 0.01

#Tiempo inicial 

t_min = 0

#Tiempo final

t_max = 1

#Vector de tiempos 
t = np.arange(t_min,t_max,dt)


xt = t
#Tamaño del vector de tiempos
L=len(t)
#Gráfico
plt.figure(1)
plt.plot(t,xt)
plt.title("Señal, función x(t)")
plt.xlabel("Tiempo (seg)")

# =============================================================================
#                       ANÁLISIS DE FRECUENCIA
# =============================================================================
#Frecuencia inicial

f_min = 0.5

#Frecuencia máxima 

f_max = 18


df=0.001

fc = np.arange(f_min, f_max, df)
recc = np.zeros(len(fc))
imcc = np.zeros(len(fc))
cc = np.zeros(len(fc))
  

recc = (1/(2 * fc * np.pi)) * (np.sin(2 * np.pi * fc)) + (1/(2 * np.pi * fc)**2)*(np.cos(2 * np.pi * fc)+1)

imcc= (1/(2 * fc * np.pi)) *(np.cos(2 * np.pi * fc)) -(1/(2 * np.pi * fc)**2)*(np.sin(2 * np.pi * fc))


a=recc*2
b=imcc*-2

cc = np.sqrt(abs(a)**2+abs(b)**2)

plt.figure(2)
plt.plot(fc,cc)

plt.title("Espectro de magnitud |x(f)|")
plt.xlabel("Frecuencia (hz)")


plt.figure(3)
plt.plot(fc,a)
plt.title('Re(x(f)) y ak')
plt.xlabel('Frecuencia (hz)')

plt.figure(4)
plt.plot(fc,b)
plt.title('Im(x(f)) y bk')
plt.xlabel('Frecuencia (hz)')

plt.figure(5)
plt.plot(fc,cc)

plt.title("|x(f)| y ck")
plt.xlabel("Frecuencia (hz)")

#Periodo de la señal
T=t_max-t_min
#Frecuecnía fundamental
fo=1/T
#Coeficiente a0 (Offset)
a0=1/2
#Definición de armónicos 
K= 16
#Incializar vectores para ak,bk,phik y ck
ak= np.zeros(2*K + 3)
bk= np.zeros(2*K + 3)
ck= np.zeros(2*K + 3)
phik= np.zeros(2*K + 3)
#Vector de frecuencia
f= np.zeros(2*K + 3)
#Cálculo de los coeficientes de Fourier
for k in range(-K-1,K+2):
    
    #Verifica signo de la componente de offset
    if a0 < 0:
        phik[k+K+1] = np.pi 
    else:
        phik[k+K+1] = 0
    
    #Caso cuando k = 0
    if k == 0:
        ak[k+K+1] = a0
        ck[k+K+1] = np.abs(a0)          
        
    else:
        ak[k+K+1] = sys.float_info.epsilon 
        #Para resolver el problema de "NAN" "no a number"
        #epsilon = 2.22 e-16 aproximado a 0
        bk[k+K+1] = -1/(k*np.pi)
        ck[k+K+1] = np.sqrt(ak[k+K+1]**2 + bk[k+K+1]**2)
        
        #Cómputo del ángulo 
        #Primer cuadrante
        if ak[k+K+1] >= 0 and bk[k+K+1] >= 0:
            phik[k+K+1] = np.arctan(bk[k+K+1]/ak[k+K+1])
        
        #Segundo cuadrante
        if ak[k+K+1] < 0 and bk[k+K+1] >= 0:
            phik[k+K+1] = np.pi - np.arctan(abs(bk[k+K+1])/np.abs(ak[k+K+1]))

        #Tercer cuadrante
        if ak[k+K+1] < 0 and bk[k+K+1] < 0:
            phik[k+K+1] = np.pi + np.arctan(np.abs(bk[k+K+1])/np.abs(ak[k+K+1]))

        #Cuarto cuadrante
        if ak[k+K+1] >= 0 and bk[k+K+1] < 0:
            phik[k+K+1] = np.arctan(abs(bk[k+K+1])/ak[k+K+1])

        #Cálculo de las Frecuencias
        f[k+K+1] = k*fo
        
        #ángulo en grados
        phik[k+K+1] = phik[k+K+1]*180/np.pi 
        
plt.figure(3)     
plt.stem(f[K+1:],ak[K+1:],"r")

plt.figure(4)     
plt.stem(f[K+1:],bk[K+1:],"r")

plt.figure(5)     
plt.stem(f[K+1:],ck[K+1:],"r")




# =============================================================================
#                   TRANSFORMADA DE FOURIER DE TIEMPO DISCRETO 
# =============================================================================
#señal discreta     
#x(t)=t, para 0<t<1
#     
#
#ITiempo muestreo

ts=0.01

fs=1/ts

#Vector de tiempos 
kts = np.arange(t_min,t_max,ts)

Xkts=kts

#Gráfico 
plt.figure(6)
plt.plot(kts,Xkts,".")
plt.title("Señal discreta")
plt.xlabel("Tiempo (seg)")


N=len(kts)

n= np.arange(0,N-1)

F=n/(N-1)

f=F*fs



df2=0.02

fd = np.arange(f_min, f_max,fs)
recd = np.zeros(len(f))
imcd = np.zeros(len(f))
cd = np.zeros(len(f))


for k in range(0, N):
  recd = recd +ts*Xkts[k]*np.cos(2*np.pi*k*F)
  imcd = imcd-ts*Xkts[k]*np.sin(2*np.pi*k*F)


ad=recd*2
bd=imcd*-2



cd = np.sqrt(abs(ad)**2+abs(bd)**2)
#f2=round(f)
fint=f.astype(int)
lim = np.where( fint==18)
lim=lim[0][0]

#Gráfico 

plt.figure(7)
plt.plot(f[:lim],cd[:lim])

plt.title("TFTD")
plt.xlabel("Frecuencia en Hz")

plt.figure(8)
plt.plot(f[:lim],cd[:lim],"blue", label="TFTD")
plt.plot(fc,cc,"red", label="TFC")
plt.title("TFTD &TFC ")
plt.xlabel("Frecuencia en Hz")
leg=plt.legend(loc="lower left")



```

#Conclusiones 

Se puede ver que los coeficientes de la serie  $a_k,~b_k~$y$~c_k$ coinciden con la parte real, imaginaria y magnitud de la transformada d Fourier, respectivamente.

Al tener una señal discreta, la transformada de Fourier discreta da una buena aproximación a los valores a los valores de la transformada de Fourier continua que nos permitira hacer diferentes tipos de análisis dependiendo la aplicación que se trabaje.
